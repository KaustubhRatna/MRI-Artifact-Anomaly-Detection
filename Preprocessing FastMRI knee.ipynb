```python
import os
import random
import numpy as np
import h5py
from tqdm import tqdm

# Paths
TRAIN_ROOT = "/kaggle/input/datasets/arafatshovon/fastmri-train-knee-2"
VAL_ROOT = "/kaggle/input/datasets/arafatshovon/fastmri-knee-val"

SAVE_ROOT = "/kaggle/working/processed_fastmri"

TRAIN_DIR = os.path.join(SAVE_ROOT, "train", "KNEE")
VAL_DIR = os.path.join(SAVE_ROOT, "val", "KNEE")

os.makedirs(TRAIN_DIR, exist_ok=True)
os.makedirs(VAL_DIR, exist_ok=True)

# 🔥 ADJUSTED
TARGET_SHAPE = (192, 192, 32)
MIN_DEPTH = 20

MIN_SLICE_MEAN = 0.01

random.seed(42)
```

```python
import os
import random

def collect_fastmri_paths(train_root, val_root):
    train_files = [os.path.join(train_root, f) for f in os.listdir(train_root) if f.endswith(".h5")]
    val_files = [os.path.join(val_root, f) for f in os.listdir(val_root) if f.endswith(".h5")]
    return train_files + val_files

all_files = collect_fastmri_paths(TRAIN_ROOT, VAL_ROOT)

# ---- REMOVE DUPLICATES (by file name) ----
unique_files = []
seen_names = set()

for f in all_files:
    name = os.path.basename(f)
    if name not in seen_names:
        unique_files.append(f)
        seen_names.add(name)

print("After dedup:", len(unique_files))

# ---- NOW shuffle and split ----
random.shuffle(unique_files)

train_files = unique_files[:450]
val_files = unique_files[450:560]

print("Train:", len(train_files))
print("Val:", len(val_files))
```

```python
def center_crop(volume, target_shape):
    x, y, z = volume.shape
    tx, ty, tz = target_shape

    start_x = max((x - tx) // 2, 0)
    start_y = max((y - ty) // 2, 0)
    start_z = max((z - tz) // 2, 0)

    return volume[
        start_x:start_x+tx,
        start_y:start_y+ty,
        start_z:start_z+tz
    ]
```

```python
def normalize(volume):
    p1, p99 = np.percentile(volume, (1, 99))

    if p99 - p1 < 1e-5:
        return None

    volume = np.clip(volume, p1, p99)
    volume = (volume - p1) / (p99 - p1)

    return volume
```

```python
def process_fastmri(paths, save_root):
    processed = 0
    skipped = 0

    for path in tqdm(paths):
        try:
            with h5py.File(path, 'r') as f:
                volume = f["reconstruction_rss"][()]  # (Z, H, W)

            # Convert to (X,Y,Z) like IXI
            volume = np.transpose(volume, (1, 2, 0))

            # Skip shallow
            if volume.shape[2] < MIN_DEPTH:
                skipped += 1
                continue

            # STEP 1: Center Crop (same as IXI)
            volume = center_crop(volume, TARGET_SHAPE)

            if volume.shape != TARGET_SHAPE:
                skipped += 1
                continue

            # STEP 2: Normalize (same)
            volume = normalize(volume)

            if volume is None:
                skipped += 1
                continue

            # 🔥 STEP 3: USE ALL SLICES (IMPORTANT CHANGE)
            slices = []

            for i in range(volume.shape[2]):
                sl = volume[:, :, i]

                if np.mean(sl) < MIN_SLICE_MEAN:
                    continue

                slices.append(sl.astype(np.float32))

            if len(slices) == 0:
                skipped += 1
                continue

            # SAVE (same structure)
            base_name = os.path.basename(path).replace(".h5", "")

            for i, sl in enumerate(slices):
                save_path = os.path.join(save_root, f"{base_name}_slice_{i:03d}.npy")
                np.save(save_path, sl)

            processed += 1

        except Exception as e:
            print(f"Error: {path} -> {e}")
            skipped += 1

    print("\nDone")
    print("Processed:", processed)
    print("Skipped:", skipped)
```

```python
print("Processing TRAIN...")
process_fastmri(train_files, TRAIN_DIR)

print("Processing VAL...")
process_fastmri(val_files, VAL_DIR)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import random

def visualize_random_samples(data_dir, num_samples=3, slices_per_sample=6):
    scan_types = ["KNEE"]   # instead of T1, T2

    for scan_type in scan_types:
        folder = os.path.join(data_dir, scan_type)

        if not os.path.exists(folder):
            print(f"{scan_type} not found")
            continue

        files = [f for f in os.listdir(folder) if f.endswith(".npy")]

        # Group slices by volume
        volume_dict = {}
        for f in files:
            base = "_".join(f.split("_slice_")[:-1])
            volume_dict.setdefault(base, []).append(f)

        volume_keys = list(volume_dict.keys())
        random.shuffle(volume_keys)

        print(f"\n===== {scan_type} =====")

        for v in volume_keys[:num_samples]:
            slice_files = volume_dict[v]

            chosen = random.sample(slice_files, min(slices_per_sample, len(slice_files)))

            fig, axes = plt.subplots(1, len(chosen), figsize=(15, 3))

            if len(chosen) == 1:
                axes = [axes]

            for i, f in enumerate(chosen):
                path = os.path.join(folder, f)
                img = np.load(path)

                axes[i].imshow(img, cmap="gray")
                axes[i].set_title(f"{f.split('_')[-1]}")
                axes[i].axis("off")

            plt.suptitle(f"{scan_type} | Volume: {v}", fontsize=12)
            plt.show()
```

```python
visualize_random_samples("/kaggle/working/processed_fastmri/train")
```

```python
import numpy as np
import os
import matplotlib.pyplot as plt
import random

def check_distribution(data_dir, num_samples=500):
    all_vals = []

    for scan_type in ["KNEE"]:
        folder = os.path.join(data_dir, scan_type)

        if not os.path.exists(folder):
            continue

        files = [f for f in os.listdir(folder) if f.endswith(".npy")]

        for f in random.sample(files, min(num_samples, len(files))):
            img = np.load(os.path.join(folder, f))
            all_vals.append(img.flatten())

    if len(all_vals) == 0:
        print("No data found!")
        return

    all_vals = np.concatenate(all_vals)

    print("Min:", np.min(all_vals))
    print("Max:", np.max(all_vals))
    print("Mean:", np.mean(all_vals))
    print("Std:", np.std(all_vals))

    plt.hist(all_vals, bins=50)
    plt.title("Pixel Intensity Distribution")
    plt.show()
```

```python
check_distribution("/kaggle/working/processed_fastmri/train")
```

```python
import os  # Added missing import

def count_unique_volumes(root):
    # Removed: from collections import set (set is a built-in)
    
    volume_ids = set()

    for split in ["train", "val"]:
        for scan_type in ["KNEE"]:
            path = os.path.join(root, split, scan_type)

            if not os.path.exists(path):
                continue

            for f in os.listdir(path):
                if f.endswith(".npy"):
                    # Extract ID before the '_slice_' delimiter
                    vol_id = f.split("_slice_")[0]
                    volume_ids.add(vol_id)

    print(f"🧠 Unique volumes used: {len(volume_ids)}")
count_unique_volumes("/kaggle/working/processed_fastmri")
```
