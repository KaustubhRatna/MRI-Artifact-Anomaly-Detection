# SimCLR — Brain MRI Anomaly Detection (IXI Dataset)

This notebook trains a **SimCLR** (Simple Contrastive Learning of Representations) model on **clean brain MRI slices** from the IXI dataset (T1 + T2 weighted) and uses the learned representations for **artifact detection via kNN anomaly scoring**.

## Approach

1. **Self-supervised pretraining**: Train a ResNet-18 encoder using the NT-Xent contrastive loss on augmented view pairs of clean MRI slices (20 epochs)
2. **Feature extraction**: Extract L2-normalized 512-d representations from the trained encoder
3. **Anomaly scoring**: Use k-nearest-neighbor (kNN) distance in the learned feature space — samples far from the clean training distribution are flagged as artifacts

## Pipeline Overview

| Stage | Details |
|-------|---------|
| **Data** | Preprocessed IXI brain slices (192×192, float32, [0,1]) — T1 and T2 |
| **Encoder** | ResNet-18 (modified for 1-channel input) → 512-d features |
| **Projection** | 512 → 512 (BN + ReLU) → 128 (used only during training) |
| **Loss** | NT-Xent with temperature τ=0.07, effective batch=256 |
| **Training** | 20 epochs, batch=64, grad accum=4, warmup=3 epochs |
| **Scoring** | kNN distance (k=5, cosine metric) on encoder features |

## 2. Data Pipeline

### Augmentations (MRI-Specific)
SimCLR relies on strong augmentations to learn meaningful representations. For medical MRI, we use **anatomically valid** transforms:
- **Random resized crop** (80–100% scale) — simulates field-of-view variation
- **Horizontal flip** only (no vertical flip for axial brain slices)
- **Rotation ±15°** — small rotations preserve anatomical structure
- **Gaussian blur** — simulates slight defocus
- **Intensity jitter** — ±10% brightness/contrast (mimics scanner variation)

> **No color augmentation** — MRI is single-channel grayscale

### Datasets
Two dataset classes:
- **`MRISimCLRDataset`** — returns two augmented views `(x1, x2)` for contrastive training
- **`MRICleanDataset`** — returns unaugmented images for **deterministic feature extraction** (critical for reliable kNN scoring)

```python
# ══════════════════════════════════════════════
# Augmentations (MRI-safe)
# ══════════════════════════════════════════════
class SimCLRAugmentMRI:
    """Two augmented views for contrastive learning.
    MRI-specific: no vertical flip, limited rotation, no color jitter."""

    def __init__(self, img_size=192):
        self.transform = transforms.Compose([
            transforms.ToTensor(),
            transforms.RandomResizedCrop(img_size, scale=(0.8, 1.0), ratio=(0.95, 1.05)),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomRotation(15),
            transforms.RandomApply([
                transforms.GaussianBlur(kernel_size=5, sigma=(0.1, 2.0))
            ], p=0.3),
            # Intensity jitter: brightness/contrast for grayscale MRI
            transforms.RandomApply([
                transforms.Lambda(lambda x: x * (0.9 + 0.2 * torch.rand(1).item()) + 0.05 * (torch.rand(1).item() - 0.5))
            ], p=0.5),
            transforms.Lambda(lambda x: torch.clamp(x, 0.0, 1.0)),
        ])

    def __call__(self, x):
        return self.transform(x), self.transform(x)


class CleanTransform:
    """No augmentation — just convert to tensor. Used for feature extraction & evaluation."""
    def __call__(self, x):
        t = torch.from_numpy(x).unsqueeze(0)  # (1, H, W)
        return t


# ══════════════════════════════════════════════
# Dataset — auto-discovers .npy files
# ══════════════════════════════════════════════
def find_npy_files(root_dir):
    """Recursively find all .npy files under root_dir.
    Returns list of (filepath, label_str) where label_str is the parent folder name."""
    paths, labels = [], []
    for dirpath, dirnames, filenames in os.walk(root_dir):
        npy_files = sorted([f for f in filenames if f.endswith(".npy")])
        if npy_files:
            folder_name = os.path.basename(dirpath)
            for f in npy_files:
                paths.append(os.path.join(dirpath, f))
                labels.append(folder_name)
    return paths, labels


class MRISimCLRDataset(Dataset):
    """Returns two augmented views (x1, x2) + scan label for contrastive training."""
    def __init__(self, root_dir, transform):
        self.transform = transform
        self.paths, label_strs = find_npy_files(root_dir)
        # Map unique folder names to integer labels (for visualization only)
        unique_labels = sorted(set(label_strs))
        self.label_map = {name: i for i, name in enumerate(unique_labels)}
        self.labels = [self.label_map[s] for s in label_strs]

        print(f"  Loaded {len(self.paths)} slices from {root_dir}")
        if unique_labels:
            print(f"  Labels: {self.label_map}")

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        img = np.load(self.paths[idx]).astype(np.float32)
        x1, x2 = self.transform(img)
        return x1, x2, self.labels[idx]


class MRICleanDataset(Dataset):
    """Returns clean (unaugmented) images + labels. Used for feature extraction."""
    def __init__(self, root_dir, transform=None):
        self.transform = transform or CleanTransform()
        self.paths, label_strs = find_npy_files(root_dir)
        unique_labels = sorted(set(label_strs))
        self.label_map = {name: i for i, name in enumerate(unique_labels)}
        self.labels = [self.label_map[s] for s in label_strs]

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, idx):
        img = np.load(self.paths[idx]).astype(np.float32)
        x = self.transform(img)
        return x, self.labels[idx]
```

## 3. Model Architecture

### Encoder (ResNet-18)
Modified for single-channel MRI input:
- `conv1`: 3-channel → **1-channel**, kernel 3×3, stride 1 (preserves spatial resolution at 192×192)
- `maxpool` removed (replaced with Identity) — avoids aggressive downsampling on small medical images
- Output: **512-dimensional** feature vector after global average pooling

### Projection Head
Maps encoder features to a 128-d space where contrastive loss is computed:
- `512 → 512` (BatchNorm + ReLU) → `512 → 128`
- BatchNorm stabilizes training with mixed precision
- **Discarded after training** — only the encoder is used for downstream anomaly scoring

```python
# ══════════════════════════════════════════════
# Model: ResNet-18 Encoder + Projection Head
# ══════════════════════════════════════════════
class Encoder(nn.Module):
    def __init__(self):
        super().__init__()
        backbone = resnet18(weights=None)
        # Adapt for 1-channel MRI input
        backbone.conv1 = nn.Conv2d(1, 64, kernel_size=3, stride=1, padding=1, bias=False)
        backbone.maxpool = nn.Identity()
        self.features = nn.Sequential(
            backbone.conv1, backbone.bn1, backbone.relu, backbone.maxpool,
            backbone.layer1, backbone.layer2, backbone.layer3, backbone.layer4,
            backbone.avgpool,
        )

    def forward(self, x):
        return self.features(x).flatten(1)  # (B, 512)


class ProjectionHead(nn.Module):
    def __init__(self, in_dim=512, hidden_dim=512, out_dim=128):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, hidden_dim),
            nn.BatchNorm1d(hidden_dim),
            nn.ReLU(inplace=True),
            nn.Linear(hidden_dim, out_dim),
        )

    def forward(self, x):
        return self.net(x)
```

## 4. NT-Xent Contrastive Loss

The **Normalized Temperature-scaled Cross-Entropy (NT-Xent)** loss:

$$\mathcal{L}_{i,j} = -\log \frac{\exp(\text{sim}(z_i, z_j) / \tau)}{\sum_{k=1}^{2N} \mathbb{1}_{[k \neq i]} \exp(\text{sim}(z_i, z_k) / \tau)}$$

where $\text{sim}(u, v) = u^\top v / (\|u\| \|v\|)$ is cosine similarity and $\tau = 0.07$ is the temperature.

> **FP16 Fix**: The loss computation is **forced to float32** via `@torch.amp.autocast(enabled=False)`. This prevents the `masked_fill_(-9e15)` overflow that crashes under half precision (fp16 max ≈ 65504). Using `float("-inf")` is safe at any precision.

```python
# ══════════════════════════════════════════════
# NT-Xent Loss (fixed for mixed precision)
# ══════════════════════════════════════════════
class NTXentLoss(nn.Module):
    def __init__(self, temperature=0.07):
        super().__init__()
        self.temperature = temperature

    @torch.amp.autocast("cuda", enabled=False)   # ← FORCE FP32 inside loss
    def forward(self, z1, z2):
        z1 = z1.float()
        z2 = z2.float()

        z1 = F.normalize(z1, dim=1)
        z2 = F.normalize(z2, dim=1)

        N = z1.size(0)
        z = torch.cat([z1, z2], dim=0)                    # (2N, D)

        sim = torch.mm(z, z.T) / self.temperature         # (2N, 2N)

        # Mask out self-similarity on diagonal
        mask = torch.eye(2 * N, dtype=torch.bool, device=z.device)
        sim.masked_fill_(mask, float("-inf"))

        # Positive pairs: (i, i+N) and (i+N, i)
        positives = torch.cat([torch.diag(sim, N), torch.diag(sim, -N)])  # (2N,)

        loss = -positives + torch.logsumexp(sim, dim=1)
        return loss.mean()
```

## 6. Model Initialization

**Optimizer**: AdamW with weight decay 1e-4 (prevents overfitting on limited MRI data)

**LR Schedule**: Linear warmup (3 epochs) → Cosine annealing to 0 over remaining 17 epochs
- Warmup prevents early divergence when the projection head is randomly initialized
- Cosine decay provides smooth convergence

**Mixed Precision**: `torch.amp.GradScaler` for faster training on GPU with automatic fp16/fp32 management.

```python
# ══════════════════════════════════════════════
# Initialize Model, Optimizer, Scheduler
# ══════════════════════════════════════════════
encoder   = Encoder().to(device)
projector = ProjectionHead().to(device)

optimizer = torch.optim.AdamW(
    list(encoder.parameters()) + list(projector.parameters()),
    lr=BASE_LR, weight_decay=WEIGHT_DECAY,
)

# Warmup + Cosine Annealing
def lr_lambda(epoch):
    if epoch < WARMUP_EPOCHS:
        return (epoch + 1) / WARMUP_EPOCHS
    progress = (epoch - WARMUP_EPOCHS) / max(EPOCHS - WARMUP_EPOCHS, 1)
    return 0.5 * (1.0 + math.cos(math.pi * progress))

scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer, lr_lambda)
criterion = NTXentLoss(temperature=TEMPERATURE)
scaler    = torch.amp.GradScaler("cuda", enabled=(device == "cuda"))

total_params = sum(p.numel() for p in encoder.parameters()) + sum(p.numel() for p in projector.parameters())
print(f"Encoder params:   {sum(p.numel() for p in encoder.parameters()):,}")
print(f"Projector params: {sum(p.numel() for p in projector.parameters()):,}")
print(f"Total:            {total_params:,}")
```

## 7. Training

**20 epochs** with batch size 64 — estimated ~3.5 hours on Kaggle T4 GPU.

**Gradient Accumulation**: Loss is divided by `GRAD_ACCUM=4` steps before backprop, and the optimizer steps every 4 mini-batches. This simulates a 256 effective batch size while fitting in GPU memory.

**Gradient Clipping**: `clip_grad_norm_(max_norm=1.0)` prevents gradient explosions, especially during early warmup when contrastive losses can spike.

**Time Tracking**: Each epoch logs wall-clock time with ETA estimate — important for Kaggle's 12h time limit.

**Checkpointing**:
- `best.pt` — saved whenever validation loss improves
- `epoch_5.pt`, `epoch_10.pt`, ... — periodic snapshots every 5 epochs
- `final.pt` — end of training (includes full history)

```python
# ══════════════════════════════════════════════
# Training & Evaluation Functions
# ══════════════════════════════════════════════
MAX_GRAD_NORM = 1.0  # gradient clipping threshold

def train_epoch(loader):
    encoder.train()
    projector.train()
    total_loss = 0.0
    optimizer.zero_grad()

    for step, (x1, x2, _) in enumerate(tqdm(loader, desc="Train")):
        x1, x2 = x1.to(device), x2.to(device)

        with torch.amp.autocast("cuda", enabled=(device == "cuda")):
            h1, h2 = encoder(x1), encoder(x2)
            z1, z2 = projector(h1), projector(h2)
            loss = criterion(z1, z2) / GRAD_ACCUM

        scaler.scale(loss).backward()

        if (step + 1) % GRAD_ACCUM == 0 or (step + 1) == len(loader):
            # Unscale before clipping, then step
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(
                list(encoder.parameters()) + list(projector.parameters()),
                MAX_GRAD_NORM,
            )
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()

        total_loss += loss.item() * GRAD_ACCUM

    return total_loss / len(loader)


@torch.no_grad()
def eval_epoch(loader):
    encoder.eval()
    projector.eval()
    total_loss = 0.0

    for x1, x2, _ in loader:
        x1, x2 = x1.to(device), x2.to(device)
        with torch.amp.autocast("cuda", enabled=(device == "cuda")):
            z1 = projector(encoder(x1))
            z2 = projector(encoder(x2))
            loss = criterion(z1, z2)
        total_loss += loss.item()

    return total_loss / len(loader)
```

```python
# ══════════════════════════════════════════════
# Training Loop
# ══════════════════════════════════════════════
history = {"train_loss": [], "val_loss": [], "lr": [], "epoch_time": []}
best_val_loss = float("inf")
total_start = time.time()

for epoch in range(EPOCHS):
    epoch_start = time.time()

    train_loss = train_epoch(train_loader)
    val_loss   = eval_epoch(val_loader)
    scheduler.step()

    epoch_time = time.time() - epoch_start
    lr_now = scheduler.get_last_lr()[0]
    history["train_loss"].append(train_loss)
    history["val_loss"].append(val_loss)
    history["lr"].append(lr_now)
    history["epoch_time"].append(epoch_time)

    elapsed = time.time() - total_start
    eta = epoch_time * (EPOCHS - epoch - 1)

    print(f"Epoch {epoch+1}/{EPOCHS}  |  train={train_loss:.4f}  val={val_loss:.4f}  "
          f"lr={lr_now:.6f}  |  {epoch_time:.0f}s/epoch  elapsed={elapsed/60:.1f}min  ETA={eta/60:.1f}min")

    # Save best model
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        torch.save({
            "epoch": epoch + 1,
            "encoder": encoder.state_dict(),
            "projector": projector.state_dict(),
            "optimizer": optimizer.state_dict(),
            "val_loss": val_loss,
        }, os.path.join(SAVE_DIR, "best.pt"))
        print(f"  ↑ New best model saved (val_loss={val_loss:.4f})")

    # Periodic checkpoint every 5 epochs
    if (epoch + 1) % 5 == 0:
        torch.save({
            "epoch": epoch + 1,
            "encoder": encoder.state_dict(),
            "projector": projector.state_dict(),
        }, os.path.join(SAVE_DIR, f"epoch_{epoch+1}.pt"))

# Save final
total_time = time.time() - total_start
torch.save({
    "epoch": EPOCHS,
    "encoder": encoder.state_dict(),
    "projector": projector.state_dict(),
    "history": history,
}, os.path.join(SAVE_DIR, "final.pt"))

print(f"\nTraining complete in {total_time/60:.1f} min ({total_time/3600:.2f} h)")
print(f"Best val loss: {best_val_loss:.4f}")
print(f"Avg epoch time: {np.mean(history['epoch_time']):.1f}s")
```

## 9. Feature Extraction & Quality Check

Extract **512-d L2-normalized features** from the encoder (projection head is discarded).

> **Important**: Features are extracted from **clean, unaugmented** images using `train_clean_loader` / `val_clean_loader`. This ensures:
> - **Deterministic** features (no random augmentation noise)
> - **All samples included** (no `drop_last` truncation)
> - **Reliable kNN scoring** — distances reflect true data similarity, not augmentation artifacts

### Collapse Detection
A critical failure mode for contrastive learning is **representation collapse** — all inputs map to the same point. We check:
- **Feature STD** (per dimension): should be > 0.01
- **Mean pairwise cosine similarity**: should be < 0.9 (lower is better, meaning features are diverse)

```python
# ══════════════════════════════════════════════
# Feature Extraction (on CLEAN unaugmented images)
# ══════════════════════════════════════════════
@torch.no_grad()
def extract_features(loader):
    """Extract L2-normalized encoder features from clean (unaugmented) images."""
    encoder.eval()
    feats, labels = [], []

    for x, y in tqdm(loader, desc="Extracting"):
        x = x.to(device)
        f = encoder(x)
        f = F.normalize(f, dim=1)
        feats.append(f.cpu().numpy())
        labels.append(y.numpy())

    return np.concatenate(feats), np.concatenate(labels)

# Use clean loaders → deterministic, unaugmented, no dropped samples
train_feats, train_labels = extract_features(train_clean_loader)
val_feats, val_labels     = extract_features(val_clean_loader)

print(f"Train features: {train_feats.shape}  (all {len(train_clean_dataset)} samples)")
print(f"Val features:   {val_feats.shape}  (all {len(val_clean_dataset)} samples)")
```

```python
# ══════════════════════════════════════════════
# Collapse Check
# ══════════════════════════════════════════════
def check_collapse(feats, name=""):
    std = np.std(feats, axis=0).mean()
    cos_sim = np.dot(feats, feats.T)
    off_diag = cos_sim[np.triu_indices_from(cos_sim, k=1)]
    mean_sim = off_diag.mean()

    print(f"{name}:")
    print(f"  Feature STD (per dim): {std:.4f}", "✅" if std > 0.01 else "⚠️ COLLAPSE")
    print(f"  Mean cosine similarity: {mean_sim:.4f}", "✅" if mean_sim < 0.9 else "⚠️ HIGH")
    return std, mean_sim

check_collapse(train_feats[:2000], "Train")
check_collapse(val_feats[:2000], "Val")
```

## 10. Anomaly Scoring (kNN Distance)

For each sample, compute the **mean cosine distance to its k=5 nearest neighbors** in the clean training feature bank:

$$\text{score}(x) = \frac{1}{k} \sum_{i=1}^{k} d_{\cos}(f(x), f(x_i^{\text{nn}}))$$

**Intuition**: Clean samples cluster tightly → low scores. Artifact-corrupted samples deviate from learned clean representations → high scores.

The score distribution over clean data establishes the "normal" baseline — at inference, samples with scores above a percentile threshold (e.g., p95) are flagged as anomalous.

```python
# ══════════════════════════════════════════════
# kNN Anomaly Scoring
# ══════════════════════════════════════════════
nn_model = NearestNeighbors(n_neighbors=K_NEIGHBORS, metric="cosine")
nn_model.fit(train_feats)

train_distances, _ = nn_model.kneighbors(train_feats)
val_distances, _   = nn_model.kneighbors(val_feats)

train_scores = train_distances.mean(axis=1)
val_scores   = val_distances.mean(axis=1)

print(f"Train scores — mean: {train_scores.mean():.4f}, std: {train_scores.std():.4f}")
print(f"Val scores   — mean: {val_scores.mean():.4f}, std: {val_scores.std():.4f}")

# Score distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].hist(train_scores, bins=60, alpha=0.7, color="steelblue", edgecolor="white")
axes[0].set_title("Train Anomaly Scores", fontsize=13)
axes[0].set_xlabel("kNN Distance")
axes[0].set_ylabel("Count")

axes[1].hist(val_scores, bins=60, alpha=0.7, color="coral", edgecolor="white")
axes[1].set_title("Val Anomaly Scores", fontsize=13)
axes[1].set_xlabel("kNN Distance")
axes[1].set_ylabel("Count")

fig.tight_layout()
fig.savefig(os.path.join(SAVE_DIR, "anomaly_scores.png"), dpi=150, bbox_inches="tight")
plt.show()
```

## 11. Embedding Visualization (t-SNE)

t-SNE projects 512-d features to 2D for visualization. Three views:
1. **Unsupervised** — reveals natural cluster structure without labels
2. **T1 vs T2** — checks if the encoder distinguishes scan modalities (expected: some separation)
3. **Anomaly score overlay** — highlights which regions of the embedding have higher kNN distances

```python
# ══════════════════════════════════════════════
# t-SNE Visualization
# ══════════════════════════════════════════════
N_TSNE = min(3000, len(val_feats))
tsne = TSNE(n_components=2, perplexity=30, random_state=SEED, n_iter=1000)
emb = tsne.fit_transform(val_feats[:N_TSNE])
labels_subset = val_labels[:N_TSNE]
scores_subset = val_scores[:N_TSNE]

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Unsupervised clusters
axes[0].scatter(emb[:, 0], emb[:, 1], s=3, alpha=0.5, c="steelblue")
axes[0].set_title("t-SNE — Unsupervised", fontsize=13)
axes[0].set_xticks([]); axes[0].set_yticks([])

# Colored by scan type (T1 vs T2)
scatter = axes[1].scatter(emb[:, 0], emb[:, 1], s=3, alpha=0.5,
                          c=labels_subset, cmap="coolwarm")
axes[1].set_title("t-SNE — T1 (blue) vs T2 (red)", fontsize=13)
axes[1].set_xticks([]); axes[1].set_yticks([])
plt.colorbar(scatter, ax=axes[1], shrink=0.8)

# Colored by anomaly score
scatter2 = axes[2].scatter(emb[:, 0], emb[:, 1], s=3, alpha=0.5,
                           c=scores_subset, cmap="YlOrRd")
axes[2].set_title("t-SNE — kNN Anomaly Score", fontsize=13)
axes[2].set_xticks([]); axes[2].set_yticks([])
plt.colorbar(scatter2, ax=axes[2], shrink=0.8)

fig.tight_layout()
fig.savefig(os.path.join(SAVE_DIR, "tsne.png"), dpi=150, bbox_inches="tight")
plt.show()
```

## 12. Feature Quality Analysis

### Cosine Similarity Distribution
A well-trained contrastive model should produce a **spread-out** similarity distribution:
- **Collapsed model**: all similarities ≈ 1.0 (everything maps to the same point)
- **Good model**: bell-shaped distribution centered well below 1.0 with reasonable spread

### Nearest Neighbor Retrieval
Qualitative check: for a random query slice, the nearest neighbors in feature space should look **visually similar** (same anatomy, contrast, slice level). This confirms the encoder captures meaningful anatomical structure.

```python
# ══════════════════════════════════════════════
# Feature Similarity Distribution
# ══════════════════════════════════════════════
subset_feats = torch.tensor(val_feats[:2000])
sim_matrix = torch.mm(subset_feats, subset_feats.T).numpy()
upper_tri = sim_matrix[np.triu_indices_from(sim_matrix, k=1)]

fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(upper_tri, bins=100, color="mediumpurple", edgecolor="white", alpha=0.8)
ax.axvline(x=upper_tri.mean(), color="red", linestyle="--", label=f"Mean: {upper_tri.mean():.3f}")
ax.set_title("Pairwise Cosine Similarity Distribution", fontsize=13)
ax.set_xlabel("Cosine Similarity")
ax.set_ylabel("Count")
ax.legend()
fig.tight_layout()
fig.savefig(os.path.join(SAVE_DIR, "similarity_dist.png"), dpi=150, bbox_inches="tight")
plt.show()
```

```python
# ══════════════════════════════════════════════
# Nearest Neighbor Retrieval (qualitative)
# ══════════════════════════════════════════════
def visualize_neighbors(dataset, feats, n_queries=3, n_neighbors=6):
    nn = NearestNeighbors(n_neighbors=n_neighbors, metric="cosine").fit(feats)
    indices = random.sample(range(len(feats)), n_queries)

    fig, axes = plt.subplots(n_queries, n_neighbors, figsize=(n_neighbors * 2.5, n_queries * 2.5))
    if n_queries == 1:
        axes = [axes]

    for row, query_idx in enumerate(indices):
        _, neighbor_idxs = nn.kneighbors([feats[query_idx]])
        for col, idx in enumerate(neighbor_idxs[0]):
            img = np.load(dataset.paths[idx])
            axes[row][col].imshow(img, cmap="gray", vmin=0, vmax=1)
            axes[row][col].axis("off")
            if col == 0:
                axes[row][col].set_title("Query", fontsize=10, fontweight="bold")
            else:
                axes[row][col].set_title(f"NN-{col}", fontsize=10)

    fig.suptitle("Nearest Neighbor Retrieval", fontsize=14, y=1.02)
    fig.tight_layout()
    fig.savefig(os.path.join(SAVE_DIR, "nn_retrieval.png"), dpi=150, bbox_inches="tight")
    plt.show()

visualize_neighbors(val_clean_dataset, val_feats)
```

## 13. Threshold Analysis & Anomaly Detection

Since all training/validation data is **clean** (no artifacts), we establish a detection threshold from the score distribution:
- Compute percentile thresholds (p90, p95, p97.5, p99) on training scores
- Apply to validation set to see how many clean samples are flagged (false positive rate)

The **p95 threshold** is saved for downstream use — at inference, any new sample scoring above this threshold is classified as an **artifact**.

### Top Anomalies
The slices with the highest kNN anomaly scores in the validation set. Since the data is clean, these represent the most "unusual" clean slices — edge cases, unusual anatomy, or boundary slices.

```python
# ══════════════════════════════════════════════
# Anomaly Detection: Score Threshold Analysis
# ══════════════════════════════════════════════

# Since we only have clean data, we analyze the score distribution
# to establish a threshold. At inference, samples above threshold = artifact.

# Percentile-based threshold from training set
for pct in [90, 95, 97.5, 99]:
    thresh = np.percentile(train_scores, pct)
    val_flagged = (val_scores > thresh).sum()
    print(f"  Threshold (train p{pct}): {thresh:.4f} → {val_flagged}/{len(val_scores)} val flagged "
          f"({100*val_flagged/len(val_scores):.1f}%)")

# Save threshold + scores for downstream use
chosen_threshold = np.percentile(train_scores, 95)
np.savez(
    os.path.join(SAVE_DIR, "scoring.npz"),
    train_scores=train_scores, val_scores=val_scores,
    train_feats=train_feats, val_feats=val_feats,
    train_labels=train_labels, val_labels=val_labels,
    threshold_p95=chosen_threshold,
)
print(f"\nScoring data saved to {SAVE_DIR}/scoring.npz")
```

```python
# ══════════════════════════════════════════════
# Top Anomalous Samples (highest kNN scores)
# ══════════════════════════════════════════════
top_k = 10
top_indices = np.argsort(val_scores)[-top_k:][::-1]

fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, idx in enumerate(top_indices):
    ax = axes[i // 5][i % 5]
    img = np.load(val_clean_dataset.paths[idx])
    ax.imshow(img, cmap="gray", vmin=0, vmax=1)
    ax.set_title(f"Score: {val_scores[idx]:.4f}", fontsize=9)
    ax.axis("off")

fig.suptitle("Top-10 Most Anomalous Validation Slices", fontsize=14)
fig.tight_layout()
fig.savefig(os.path.join(SAVE_DIR, "top_anomalies.png"), dpi=150, bbox_inches="tight")
plt.show()

print("All outputs saved to:", SAVE_DIR)
print("Files:", os.listdir(SAVE_DIR))
```
